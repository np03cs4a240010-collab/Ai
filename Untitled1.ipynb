```python
import numpy as np

random_array = np.random.random((10, 10))

min_value = random_array.min()
max_value = random_array.max()

print("Random 10x10 Array:\n", random_array)
print("\nMinimum value:", min_value)
print("Maximum value:", max_value)
```

```python
import numpy as np

arr = np.ones((5, 5), dtype=int)

# Set the inner elements to 0
arr[1:-1, 1:-1] = 0

print(arr)
```

```python
import numpy as np

# Create an 8x8 matrix filled with zeros
checkerboard = np.zeros((8, 8), dtype=int)

# Fill with checkerboard pattern
checkerboard[1::2, ::2] = 1   # rows 1,3,5,7 → columns 0,2,4,6
checkerboard[::2, 1::2] = 1   # rows 0,2,4,6 → columns 1,3,5,7

print(checkerboard)
```
