3.2 Exploratory Data Analysis

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

Which passenger had the highest fare paid according to their age?

```python
#1
df['fare_per_year'] = df['Fare'] / df['Age']
df.head()
```

```python
#2
high_fare_age = df[df['fare_per_year'] > 5]
high_fare_age.head()
```

```python
#3
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
high_fare_age_srt.head()
```

```python
#4
result = high_fare_age_srt[['Name', 'fare_per_year']]
result.head()
```

Which adult male passenger (age  18 and Sex is male) paid the highest fare relative to their class?

```python
#1
df['fare_per_class'] = df['Fare'] / df['Pclass']
```

```python
#2
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]
```

```python
#3
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)
```

```python
#4
result2 = adult_males_srt[['Name', 'Age', 'fare_per_class']]
result2.head()
```

3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

What percent of the total fare revenue came from each passenger class?

```python
#1
total_fare = df['Fare'].sum()
total_fare
```

```python
#2
fare_1 = df[df['Pclass'] == 1]['Fare'].sum()
fare_1
```

```python
#3
fare_2 = df[df['Pclass'] == 2]['Fare'].sum()
fare_2
```

```python
#4
fare_3 = df[df['Pclass'] == 3]['Fare'].sum()
fare_3
```

```python
#5
fare_list = [fare_1, fare_2, fare_3]
proportion = [x / total_fare for x in fare_list]
proportion
```

What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?

```python
#1
def age_group(x):
    if x < 18:
        return "child"
    elif x < 65:
        return "adult"
    else:
        return "senior"

df['age_group'] = df['Age'].apply(age_group)

df[['Name', 'Age', 'age_group']].head()
```

```python
#2
total_passengers = len(df)
total_passengers
```

```python
#3
count_groups = df['age_group'].value_counts()
count_groups
```

```python
#4
percentage = (count_groups / total_passengers) * 100
percentage
```
